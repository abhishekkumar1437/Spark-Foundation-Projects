```python
import numpy as np
import pandas as pd
```

```python
data= pd.read_csv('./WHInterns.csv')
data.head(2)
```

```python
sample=pd.read_csv('./SIB_sample.csv')
sample
```

```python

data['EMAIL']=data['Email Address']
firstName=[]
lastName=[]
for name in data['Full Name']:
  names=name.split(' ')
  firstName.append(names[0])
  if(len(names)>=2):
    lastName.append(names[1])
  else:
    lastName.append("")
data['FIRSTNAME']=firstName
data['LASTNAME']=lastName
data['PHONE']=data['Phone']
data['CREATION_DATE']=data['Timestamp']
gender=[]
for ge in data['Gender']:
  if ge=='Male':
    gender.append(1)
  else:
    gender.append(2)
data['GENDER']=gender
data.head(3)
```

```python
print(data.head(5))
data=data.drop(['Timestamp','Full Name','Email Address','Gender','DOB','Phone','Apply for','Do you have prior experience?','College Name','LinkedIn Profile URL','Resume / CV'],axis=1)
```
